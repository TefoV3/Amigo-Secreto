```python
import smtplib
import pandas as pd
import random
# Parte 1
name_account="Andres Ortega" # nombre de la cuenta
email_account="***@gmail.com" # correo de la cuenta a utilizar
password_account= "****************" # clave de aplicacion del paso 1 ! quitar espacios

# Parte 2
import smtplib
#'smtp.gmail.com' y puerto 465 corresponde solo a Gmail
#'smtp-mail.outlook.com' y puerto 587
server=smtplib.SMTP_SSL('smtp.gmail.com', 465)
#Saludo al servidor
server.ehlo()
# Inicio de sesion
server.login(email_account, password_account)

#Parte 3
import pandas as pd
email_df = pd.read_excel(r'C:\Users\estef\OneDrive\Escritorio\Productivo\ProyectosVisual\AmigoSecreto\amigo_secreto.xlsx') # leo el excel
#obtengo nombre
all_name= email_df['Name'] 

# Generar asignación aleatoria sin que coincida consigo mismo
def generar_asignacion(names):
    asignacion = names.copy()
    while True:
        random.shuffle(asignacion)
        # Verificar que nadie se asigne a sí mismo
        if all(asignacion[i] != names[i] for i in range(len(names))):
            return asignacion

all_names_aleatorio = generar_asignacion(all_name)

#obtengo correos
all_emails= email_df["Email"]
#obtener asuntos
all_subjects= email_df["Subject"]
#Mensajes
all_messages=email_df['Message']

print(all_name)
print('*************')
print(all_names_aleatorio)


#Parte 4
for i in range(len(email_df)): # genero el correo
    name = all_name[i]
    email = all_emails[i]
    subject= all_subjects[i]
    message=('Hola ' + all_name[i] +'\n\n' + f'Tu amigo secreto es: {all_names_aleatorio[i]} CTM!!! (1$ DE REGALO)')
    
    # generar el formato de enviar

    sent_email=("From: {0} <{1}> \n"
                "To: {2} <{3}> \n"
                "Subject: {4} \n\n"
                "{5}".format(name_account,email_account,name,email,subject,message)
                )
    try: 
        server.sendmail(email_account, [email], sent_email)
    except Exception as e:
        print(f"no se pudo enviar, error: {e}")
server.close()
```
